```python
import pandas as pd
```

```python
meteo2021 = pd.read_pickle('df_meteo_2021.h5') # ci mette davvero poco
```

```python
meteo2021['anno'] = pd.DatetimeIndex(meteo2021['Data']).year.astype(str)
meteo2021['mese'] = pd.DatetimeIndex(meteo2021['Data']).month.astype(str)
meteo2021['giorno'] = pd.DatetimeIndex(meteo2021['Data']).day.astype(str)
meteo2021.head()
```

```python
meteo2021['data2'] = meteo2021['anno'] + " " + meteo2021['mese'] + " " + meteo2021['giorno']
meteo2021.head()
```

```python
del meteo2021['anno']
del meteo2021['mese']
del meteo2021['giorno']
```

```python
meteo2021.head()
```

```python
meteo2021_grouped = meteo2021.groupby(["data2","IdSensore"]).sum()
```

```python
meteo2021_grouped.index.names = ['Data','IdSensore']
```

```python
meteo2021_grouped.shape
```

```python
meteo2021_grouped.head()
```

```python
meteo2021_grouped.to_pickle('df_meteo_2021.h5') 
```
