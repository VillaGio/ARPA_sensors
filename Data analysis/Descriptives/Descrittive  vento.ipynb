correlation matrix

```python
df_corr = pd.DataFrame()

appo = grouped2021['Valore']
appo.index = range(len(appo))
df_corr["2021"] = appo

appo = pd.concat([grouped2020['Valore'][:59],grouped2020['Valore'][60:]])
appo.index = range(len(appo))
df_corr["2020"] = appo

appo = grouped2019['Valore']
appo.index = range(len(appo))
df_corr["2019"] = appo

appo = grouped2018['Valore']
appo.index = range(len(appo))
df_corr["2018"] = appo

appo = grouped2017['Valore']
appo.index = range(len(appo))
df_corr["2017"] = appo

appo = pd.concat([grouped2016['Valore'][:59],grouped2016['Valore'][60:]])
appo.index = range(len(appo))
df_corr["2016"] = appo

appo = grouped2015['Valore']
appo.index = range(len(appo))
df_corr["2015"] = appo

appo = grouped2014['Valore']
appo.index = range(len(appo))
df_corr["2014"] = appo

appo = grouped2013['Valore']
appo.index = range(len(appo))
df_corr["2013"] = appo

appo = pd.concat([grouped2012['Valore'][366:424],grouped2012['Valore'][425:]])
appo.index = range(len(appo))
df_corr["2012"] = appo

appo = grouped2012['Valore'][:365]
appo.index = range(len(appo))
df_corr["2011"] = appo

import seaborn as sn


rcParams['figure.figsize'] = 15,8

corrMatrix = df_corr.corr()
#sn.color_palette("flare", as_cmap=True)
sn.heatmap(corrMatrix, annot=True, vmin=0, vmax=1, cmap="flare").set(title='Correlation matrix between yearly wind speed time series')
plt.show()
```

Confronto tra serie storiche

```python
plt.plot(df_corr["2021"],  label = "2021", color='lawngreen')
plt.plot(df_corr["2011"], label = "2011", color='turquoise')
plt.title("Comparison between wind speed measurements in 2011 and 2021")
#plt.hlines(y=50, xmin=0, xmax=365, colors='red', linestyles='-', lw=2, label='PM10 daily threshold')
plt.ylabel("m/s")
plt.xlabel("nth day of the year")
plt.legend()
plt.show()
```
