## Line plots

For each years (= dataframe) extract valid (> 0) records concerning PM10. Group them accoriding to the date, using mean as aggregation function:

### PM10 lineplot

```python
d2021pm10 = pd.merge(left=data2021[data2021['valore']>0], right=stazioni[(stazioni['NomeTipoSensore']=="PM10") | (stazioni['NomeTipoSensore']=="PM10 (SM2005)")], how="inner")
d2021pm10['dataora']= pd.to_datetime(d2021pm10.dataora).dt.date
d2021pm10 = d2021pm10.groupby(["dataora"]).mean()
pm10_2021_plot = d2021pm10["valore"]

plt.plot(pm10_2021_plot, marker='o')
plt.title("2021 average measurement of PM10 in Lombardy")
plt.hlines(y=50, xmin=18620, xmax=18620+380, colors='red', linestyles='-', lw=2, label='PM10 daily threshold')
plt.ylabel("concentration in µg/m3")
plt.xlabel("Date")
plt.rcParams["figure.figsize"] = (15,6)
plt.legend()
plt.show()
```
